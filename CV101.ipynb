### Conversion from RGB to HSV color space

```python
def rgb_hsv(r,g,b,sf):
    '''can change the scalefactor for s and v to range between 0 and some number instead of 
    0 and 1 by providing the value for the scale factor'''
    r/=255.0 # normalize the rgb value
    g/=255.0
    b/=255.0
    cmax = max(r,g,b) # get the maximum of r g and b
    cmin = min(r,g,b) # get the minimum of r g abd b
    diff = cmax-cmin
    if cmax==cmin:
        h = 0
    elif cmax==r:
        h = (60*(0+(g-b)/diff))%360 # to make the value between 0 and 360 if incase it is not
    elif cmax==g:
        h = (60*(2+(b-r)/diff))%360
    elif cmax==b:
        h = (60*(4+(r-g)/diff))%360
        
    if h<0:
        h+=360
    if cmax==0:
        s=0
    
    else:
        s = (diff/cmax)*sf
    v = cmax*sf
    
    return h,s,v
```

```python
print(rgb_hsv(100,200,50,100))
```

```python
## extracting red color flower
```

```python
img = cv2.imread('./images/tulip.jpeg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```

```python
lower_red = np.array([0,120,70])
upper_red = np.array([15,255,255])

mask_1 = cv2.inRange(hsv,lower_red,upper_red) # gives us the binaray mask
```

```python
type(mask_1)
```

```python
mask_1.dtype
```

```python
mask_1.shape
```

```python
mask_1.max()
```

```python
mask_1.min()
```

```python
# pixels that are in the range of lower_red and upper_red range is picked while the pixels that are outside the range
# are not picked
```

```python
lower_red = np.array([170,120,70])
upper_red = np.array([180,255,255])

mask_2 = cv2.inRange(hsv,lower_red,upper_red) # gives us the binaray mask
```

```python
np.unique(mask_2)
```

```python
# there are 2 unique values,, 0 means not present while 255 means present
```

```python
np.unique(mask_1)
```

```python
# bitwise or
mask = mask_1 | mask_2
```

```python
np.unique(mask)
```

```python
res_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res_img[:,:,::-1])
plt.show()
```
