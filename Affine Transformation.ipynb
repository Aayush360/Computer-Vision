```python
import matplotlib as mpl


def dispalyImgInActualSize(img):
    dpi = mpl.rcParams['figure.dpi']
    H,W = img.shape # num_rows,num_cols
    
    # What size does the figure need to be in inches to fit the image?
    figsize = W/float(dpi), H/float(dpi) #gives figure size in inches
    
    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize = figsize)
    ax = fig.add_axes([0,0,1,1]) # fitting original image into the axes
    
    # Hide spines, ticks, etc.
    ax.axis('off')
    
    # display the image
    ax.imshow(img,cmap='gray')
    plt.show
```

```python
def bilinear(r,c,img):
    '''
    r: candidate row location
    c: candidate column location
    img: grayscale source image
    
    '''
    
    # check for boundary location
    
    
    lc = int(c)
    rc = lc+1

    tr = int(r)
    br = tr+1

    # weight for right and left
    r_wt = c-lc
    l_wt = rc-c

    # weight for top and bottom
    t_wt = br-r
    b_wt = r-tr
    
    if tr>=0 and br< img.shape[0] and lc>=0 and rc< img.shape[1]:

        a = l_wt*img[tr,lc]+r_wt*img[tr,rc]

        b = l_wt*img[br,lc]+r_wt*img[br,rc]

    # final intensity
        c = a*t_wt+b*b_wt

        return np.uint8(c)
   
    else:
        
        return 0    
        
```

```python
def get_extents_affine(img,trans_mat):
    '''
    input:
    -----------------------------------------------------
    trans_mat: any general transformation matrix e.g 
    
    rotation matrix that does the rotation
    
    output:
    -------------------------------------------------------
    returns the height and width of the output image coordinates
    and the blank output image with the required shape for affine transformation

    '''
    
    num_rows = img.shape[0]
    num_cols = img.shape[1]
    
    a = np.array([0,0,1])
    b = np.array([num_cols-1,0,1])
    c = np.array([num_cols-1,num_rows-1,1])
    d = np.array([0,num_rows-1,1])
    
    new_coord = np.array([a,b,c,d])
    
#     print(original_coord.shape)
    trans_coord=new_coord.dot(trans_mat)
    
    min_col = trans_coord.min(axis=0)[0]
    min_row = trans_coord.min(axis=0)[1]
    max_col = trans_coord.max(axis=0)[0]
    max_row = trans_coord.max(axis=0)[1]
    
    # rounding off the values
    
    min_col = np.int64(np.floor(min_col))
    min_row = np.int64(np.floor(min_row))
    max_col = np.int64(np.ceil(max_col))
    max_row = np.int64(np.ceil(max_row))
    
    
    H = max_row-min_row+1
    W = max_col-min_col+1
    
    
    out_img = np.zeros((H,W), dtype='uint8')
    
    return min_row,max_row,min_col,max_col,out_img,H,W
```

```python
# the same rotation matrix is capable of performing all the transformations if given the appropriate matrix

def affine_transform(img,trans_mat):
    '''
    
    input:
    --------------------------------------------------
    img: source image to be transformed
    img_res = resultant blank image where output of the source image should be mapped to
    trans_mat: transformation matrix
    max_rows, min_rows, max_cols, min_cols are the maximum and minimum values of the image coordinates
    
    output:
    ---------------------------------------------------
    Displays the transformed image
    
    
    '''
    num_rows = img.shape[0]
    num_cols = img.shape[1]
    
    min_row,max_row,min_col,max_col,out_img,H,W=get_extents_affine(img,trans_mat)
    Tinv = np.linalg.inv(trans_mat)

    for new_i in range(min_row,max_row):
        for new_j in range(min_col,max_col):
            p_dash = np.array([new_i,new_j,1])
            p = Tinv.dot(p_dash) # find corresponding coordinate in original image from the transformed image
            i,j = p[0],p[1]
#             print(i,j)

            if i<0 or i>=num_rows or j <0 or j>=num_cols:
                pass
            else:
                g=bilinear(i,j,img)
                out_img[new_i-min_row,new_j-min_col]=g
    dispalyImgInActualSize(out_img)
    return out_img
```

### Reflection about any Line

To apply reflection about any arbirary line, we first need to translate the line such that it passes through the origin. We then need to rotate it such that it is inclined to the x-axis/y-axis. finally we can perform the reflection about that particular axis (x-axis/y-axis).

The same process need to be applied in reverse i.e rotate using the inverse of the applied rotation matirx and perform the inverse of the translation matrix

```python
# say we need to perfrom reflection about the line y=x+2 (r=c+2)
```

```python
T_mat = np.array([[1,0,-2],[0,1,0],[0,0,1]])

T_inv_mat = np.linalg.inv(T_mat)
```

### Projective Transformation

```python
def get_extents_projective(img,trans_mat):
    '''
    input:
    -----------------------------------------------------
    trans_mat: any general transformation matrix e.g 
    
    rotation matrix that does the rotation
    
    output:
    -------------------------------------------------------
    returns the height and width of the output image coordinates
    and the blank output image with the required shape for affine transformation

    '''
    
    num_rows = img.shape[0]
    num_cols = img.shape[1]
    
    a = np.array([0,0,1])
    b = np.array([num_cols-1,0,1])
    c = np.array([num_cols-1,num_rows-1,1])
    d = np.array([0,num_rows-1,1])
    
    new_coord = np.array([a,b,c,d])
    
#     print(original_coord.shape)
    trans_coord=new_coord.dot(trans_mat)
    trans_coord = trans_coord/ trans_coord[:,-1:]
    
    min_col = trans_coord.min(axis=0)[0]
    min_row = trans_coord.min(axis=0)[1]
    max_col = trans_coord.max(axis=0)[0]
    max_row = trans_coord.max(axis=0)[1]
    
    # rounding off the values
    
    min_col = np.int64(np.floor(min_col))
    min_row = np.int64(np.floor(min_row))
    max_col = np.int64(np.ceil(max_col))
    max_row = np.int64(np.ceil(max_row))
    
    
    H = max_row-min_row+1
    W = max_col-min_col+1
    
    
    out_img = np.zeros((H,W), dtype='uint8')
    
    return min_row,max_row,min_col,max_col,out_img,H,W
```

```python
# the same rotation matrix is capable of performing all the transformations if given the appropriate matrix

def projective_transform(img,trans_mat):
    '''
    
    input:
    --------------------------------------------------
    img: source image to be transformed
    img_res = resultant blank image where output of the source image should be mapped to
    trans_mat: transformation matrix
    max_rows, min_rows, max_cols, min_cols are the maximum and minimum values of the image coordinates
    
    output:
    ---------------------------------------------------
    Displays the transformed image
    
    
    '''
    num_rows = img.shape[0]
    num_cols = img.shape[1]
    
    min_row,max_row,min_col,max_col,out_img,H,W=get_extents_projective(img,trans_mat)
    Tinv = np.linalg.inv(trans_mat)

    for new_i in range(min_row,max_row):
        for new_j in range(min_col,max_col):
            p_dash = np.array([new_i,new_j,1])
            p = Tinv.dot(p_dash) # find corresponding coordinate in original image from the transformed image
            p = p/p[2]
            i,j = p[0],p[1]
#             print(i,j)

            if i<0 or i>=num_rows or j <0 or j>=num_cols:
                pass
            else:
                g=bilinear(i,j,img)
                out_img[new_i-min_row,new_j-min_col]=g
    dispalyImgInActualSize(out_img)
    return out_img
```
